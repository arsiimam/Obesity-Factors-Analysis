# **Distribution Visualization**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('../Data/Obesity_DataSet_raw.csv')
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

## **Distribution Visualization Using Histogram**

```python
# Select the numeric columns in the dataset
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Define the number of columns and rows for the subplot grid
num_cols = 2
num_rows = (len(numeric_cols) + num_cols - 1) // num_cols  # Calculate the required number of rows

# Set the figure size to fit all subplots
plt.figure(figsize=(15, num_rows * 5))

# Iterate through each numeric column and create a histogram
for i, column in enumerate(numeric_cols, 1):
    plt.subplot(num_rows, num_cols, i)  # Create a subplot for each numeric column
    sns.histplot(df[column], kde=True, color='skyblue')  # Create the histogram with a KDE (Kernel Density Estimate) for smoother distribution
    plt.title(f"Distribution: {column}")
    plt.xlabel(column)
    plt.ylabel('Density')

# Adjust layout to avoid overlap between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

## **Distribution Visualization Using Boxplots**

```python
# Select numeric columns from the dataframe
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Set the number of columns for subplots and calculate the number of rows
num_cols = 2
num_rows = (len(numeric_cols) + num_cols - 1) // num_cols 

# Create a figure with the appropriate size
plt.figure(figsize=(15, num_rows * 5))

# Create a box plot for each numeric column
for i, column in enumerate(numeric_cols, 1):
    plt.subplot(num_rows, num_cols, i) 
    sns.boxplot(df[column], color='skyblue')
    plt.title(f"Distribution: {column}")
    plt.xlabel(column)
    plt.ylabel('Density')

# Adjust layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

## **Distribution Visualization Using Violon Plots**

```python
# Select the numeric columns in the dataset
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Define the number of columns and rows for the subplot grid
num_cols = 2
num_rows = (len(numeric_cols) + num_cols - 1) // num_cols  # Calculate the required number of rows

# Set the figure size to fit all subplots
plt.figure(figsize=(15, num_rows * 5))

# Iterate through each numeric column and create a vertical violin plot
for i, column in enumerate(numeric_cols, 1):
    plt.subplot(num_rows, num_cols, i)  # Create a subplot for each numeric column
    sns.violinplot(y=df[column], color='skyblue')  # Create the vertical violin plot (using y instead of x)
    plt.title(f"Distribution: {column}")
    plt.xlabel('Density')
    plt.ylabel(column)

# Adjust layout to avoid overlap between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
